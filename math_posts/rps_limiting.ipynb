```python
lmbd = target_rps = 129.01 + 7.27 
```

```python
def get_prob_at_least(border, lmbd):
    sum = 1
    mult = 1
    for i in range(1, border):
        mult = mult * lmbd / i
        sum += mult
    return 1 - sum * math.exp(-lmbd)
```

```python
get_prob_at_least(151, lmbd)
```

```python
def get_expectation_via_condtional_at_least(border, lmbd):
    sum_prob = 0
    exp_sum = 0
    mult = 1
    for i in range(1, 10000):
        mult = mult * lmbd / i
        if (i >= border):
            exp_sum += i * mult
            sum_prob += mult
    return exp_sum / sum_prob
```

```python
def get_result(events, border):
    return (get_expectation_via_condtional_at_least(border, events) - border) * get_prob_at_least(border, events)
```
